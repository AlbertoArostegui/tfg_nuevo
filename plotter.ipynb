```python
import matplotlib.pyplot as plt
import numpy as np

gpt_4o = {
          "Inicial": [.74, .73, .69],
          "Grafos de conocimiento": [.55, .58, .6],
          "Embeddings": [.61, .57, .58],
          "Ontología": [.72, .7, .75]
         }

gpt_3_5_turbo = {"Inicial": [.6, .68, .6],
                 "Grafos de conocimiento": [.49, .52, .58],
                 "Embeddings": [.64, .6, .59],
                 "Ontología": [.55, .6, .59]
                 }
# Calculate averages
gpt_4o_avg = {k: np.mean(v) for k, v in gpt_4o.items()}
gpt_3_5_turbo_avg = {k: np.mean(v) for k, v in gpt_3_5_turbo.items()}

# Calculate standard deviations
gpt_4o_std = {k: np.std(v) for k, v in gpt_4o.items()}
gpt_3_5_turbo_std = {k: np.std(v) for k, v in gpt_3_5_turbo.items()}

# Plotting
approaches = list(gpt_4o.keys())
x = np.arange(len(approaches))  # label locations
width = 0.35  # width of the bars

fig1, ax1 = plt.subplots(figsize=(8, 6))
ax1.bar(x, gpt_3_5_turbo_avg.values(), width, label='GPT-3.5-turbo', yerr=gpt_3_5_turbo_std.values(), color='lightcoral')
ax1.set_ylabel('Scores')
ax1.set_title('Precisión por propuesta GPT-3.5-turbo')
ax1.set_xticks(x)
ax1.set_xticklabels(approaches)
fig1.tight_layout()
plt.show()

# GPT-4o
fig2, ax2 = plt.subplots(figsize=(8, 6))
ax2.bar(x, gpt_4o_avg.values(), width, label='GPT-4o', yerr=gpt_4o_std.values(), color='lightblue')
ax2.set_ylabel('Scores')
ax2.set_title('Precisión por propuesta GPT-4o')
ax2.set_xticks(x)
ax2.set_xticklabels(approaches)
fig2.tight_layout()
plt.show()
```
