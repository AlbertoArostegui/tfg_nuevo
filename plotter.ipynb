```python
import matplotlib.pyplot as plt
import numpy as np

gpt_4o = {
          "Inicial": [.69, .71, .65],
          "Grafos de conocimiento": [.6, .61, .56],
          "Embeddings": [.73, .69, .68],
          "Ontologa": [.74, .7, .71]
         }

gpt_3_5_turbo = {"Inicial": [.46, .5, .5],
                 "Grafos de conocimiento": [.45, .44, .47],
                 "Embeddings": [.56, .57, .59],
                 "Ontologa": [.59, .63, .60]
                 }
# Calculate averages
gpt_4o_avg = {k: np.mean(v) for k, v in gpt_4o.items()}
gpt_3_5_turbo_avg = {k: np.mean(v) for k, v in gpt_3_5_turbo.items()}

# Calculate standard deviations
gpt_4o_std = {k: np.std(v) for k, v in gpt_4o.items()}
gpt_3_5_turbo_std = {k: np.std(v) for k, v in gpt_3_5_turbo.items()}

# Plotting
approaches = list(gpt_4o.keys())
x = np.arange(len(approaches))  # label locations
width = 0.35  # width of the bars
def autolabel(rects, ax):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(round(height, 2)),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, -20),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

fig1, ax1 = plt.subplots(figsize=(8, 6))
rects1 = ax1.bar(x, gpt_3_5_turbo_avg.values(), width, label='GPT-3.5-turbo', yerr=gpt_3_5_turbo_std.values(), color='lightcoral')
ax1.set_ylabel('Scores')
ax1.set_title('Precisin por propuesta GPT-3.5-turbo')
ax1.set_xticks(x)
ax1.set_xticklabels(approaches)
ax1.set_ylim(0, 1)
autolabel(rects1, ax1)
fig1.tight_layout()
plt.show()

fig2, ax2 = plt.subplots(figsize=(8, 6))
rects2 = ax2.bar(x, gpt_4o_avg.values(), width, label='GPT-4o', yerr=gpt_4o_std.values(), color='lightblue')
ax2.set_ylabel('Scores')
ax2.set_title('Precisin por propuesta GPT-4o')
ax2.set_xticks(x)
ax2.set_xticklabels(approaches)
ax2.set_ylim(0, 1)
autolabel(rects2, ax2)
fig2.tight_layout()
plt.show()
```
